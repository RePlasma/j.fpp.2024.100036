```python
import numpy as np
from numpy import log, log10, sin, cos, exp, pi, sqrt
from scipy.special import gamma
import matplotlib.pyplot as plt
import json
```

### Figure 1

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
f = open('dataWPD/fig1/wpd.json')
data = json.load(f)
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'k.',markersize=1,label=r'paper')

px = np.linspace(0,6,200)
pb0 = 3; Dp0 = 1
def f44(px,m):
    return 1/(2*gamma(1+1/m)*Dp0) * exp( -((px-pb0)/Dp0)**m )
plt.plot(px, f44(px,2), '-', alpha=0.7, label=r'eq 44')
plt.plot(px, f44(px,4), '-', alpha=0.7)
plt.plot(px, f44(px,6), '-', alpha=0.7)
plt.plot(px, f44(px,8), '-', alpha=0.7)
#
plt.title(r'Figure 1')
plt.xlim([0,6])
plt.xlabel(r'$p_x$')
plt.ylabel(r'$f(p_x,t)$')
xmin, xmax = axes.get_xlim()
ymin, ymax = axes.get_ylim()
axesratio = (ymax-ymin)/(xmax-xmin)
axes.set_aspect(aspect=3/4/axesratio)
axes.legend()
plt.tight_layout()
plt.show()
```

### Figure 2

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
f = open('dataWPD/fig2/wpd.json')
data = json.load(f)
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'k.',markersize=1,label=r'paper')

px = np.linspace(0,6,200)
pb0 = 3; Dp0 = 1
def f46(px,s1t,m=8):
    P1=1
    return 1/(2*gamma(1+1/m)*Dp0*P1*(1-px*s1t)**2) * exp( -((px-pb0*(1-px*s1t))/Dp0/(1-px*s1t))**m )
plt.plot(px, f46(px,0), '-', alpha=0.7, label=r'eq 46')
plt.plot(px, f46(px,0.1), '-', alpha=0.7)
plt.plot(px, f46(px,0.2), '-', alpha=0.7)
plt.plot(px, f46(px,0.3), '-', alpha=0.7)
plt.plot(px, f46(px,0.4), '-', alpha=0.7)
plt.plot(px, f46(px,0.5), '-', alpha=0.7)
#
plt.title(r'Figure 2')
plt.xlim([0,4.5])
plt.xlabel(r'$p_x$')
plt.ylabel(r'$f(p_x,t)$')
xmin, xmax = axes.get_xlim()
ymin, ymax = axes.get_ylim()
axesratio = (ymax-ymin)/(xmax-xmin)
axes.set_aspect(aspect=3/4/axesratio)
axes.legend()
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
f = open('dataWPD/fig4/wpd.json')
data = json.load(f)
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'k.',markersize=1,label=r'paper')

px = np.linspace(0.1,12,200)
pb0 = 10; Dp0 = 1.0
def f49(px,s2t,m=8):
    P2=1
    return 1/(2*gamma(1+1/m)*Dp0*P2) * (1+s2t/(px)**(1/3))**2 * exp( -( ((px**(1/3)+s2t)**3-pb0)/Dp0 )**m )
plt.plot(px, f49(px,0), '-', alpha=0.7, label=r'eq 49')
plt.plot(px, f49(px,0.1), '-', alpha=0.7)
plt.plot(px, f49(px,0.2), '-', alpha=0.7)
plt.plot(px, f49(px,0.3), '-', alpha=0.7)
plt.plot(px, f49(px,0.4), '-', alpha=0.7)
plt.plot(px, f49(px,0.5), '-', alpha=0.7)
#
plt.title(r'Figure 4')
plt.xlim([0,12])
plt.xlabel(r'$p_x$')
plt.ylabel(r'$f(p_x,t)$')
xmin, xmax = axes.get_xlim()
ymin, ymax = axes.get_ylim()
axesratio = (ymax-ymin)/(xmax-xmin)
axes.set_aspect(aspect=3/4/axesratio)
axes.legend()
plt.tight_layout()
plt.show()
```

### Figure 5

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
f = open('dataWPD/fig5/wpd.json')
data = json.load(f)
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'k.',markersize=1,label=r'paper')

px = np.linspace(0.1,3,200)
pm0 = 2.5; Dp0 = 1.0
def f54(px,s1t):
    P1=1
    argg = pm0**2 * (1-px*s1t)**2 - px**2
    if argg > 0:
        return 4/( pi*P1*(1-px*s1t) * sqrt( argg ) ) # lacking a factor of x2
    else:
        return 0
f54 = np.vectorize(f54)
plt.plot(px, f54(px,0), '-', alpha=0.7, label=r'eq 54')
plt.plot(px, f54(px,0.1), '-', alpha=0.7)
plt.plot(px, f54(px,0.2), '-', alpha=0.7)
plt.plot(px, f54(px,0.3), '-', alpha=0.7)
plt.plot(px, f54(px,0.4), '-', alpha=0.7)
plt.plot(px, f54(px,0.5), '-', alpha=0.7)
#
plt.title(r'Figure 5')
plt.xlim([0,3])
plt.xlabel(r'$p_x$')
plt.ylabel(r'$f(p_x,t)$')
xmin, xmax = axes.get_xlim()
ymin, ymax = axes.get_ylim()
axesratio = (ymax-ymin)/(xmax-xmin)
axes.set_aspect(aspect=3/4/axesratio)
axes.legend()
plt.tight_layout()
plt.show()
```

### Figure 6

```python
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(5,4))
f = open('dataWPD/fig6/wpd.json')
data = json.load(f)
datatmp = data['datasetColl'][0]['data']
datatmp2 = np.array([datatmp[i]['value'] for i in range(len(datatmp))])
indxs = np.argsort(datatmp2[:,0])
plt.plot(datatmp2[:,0][indxs],datatmp2[:,1][indxs],'k.',markersize=1,label=r'paper')

px = np.linspace(0.1,14.9,200)
pm0 = 15; Dp0 = 1.0
def f56(px,s2t):
    P2=1
    argg = pm0**2  - (px**(1/3) + s2t)**6
    if argg > 0:
        return 2*(1+s2t * px**(-1/3))**2 / (pi*P2 * sqrt(argg)) 
    else:
        return 0
f56 = np.vectorize(f56)
plt.plot(px, f56(px,0), '-', alpha=0.7, label=r'eq 56')
plt.plot(px, f56(px,0.1), '-', alpha=0.7)
plt.plot(px, f56(px,0.2), '-', alpha=0.7)
plt.plot(px, f56(px,0.3), '-', alpha=0.7)
plt.plot(px, f56(px,0.4), '-', alpha=0.7)
plt.plot(px, f56(px,0.5), '-', alpha=0.7)
plt.plot(px, f56(px,0.6), '-', alpha=0.7)
#
plt.title(r'Figure 6')
plt.xlim([0,15])
plt.ylim([0,3])
#plt.yscale('log')
plt.xlabel(r'$p_x$')
plt.ylabel(r'$f(p_x,t)$')
xmin, xmax = axes.get_xlim()
ymin, ymax = axes.get_ylim()
axesratio = (ymax-ymin)/(xmax-xmin)
axes.set_aspect(aspect=3/4/axesratio)
axes.legend()
plt.tight_layout()
plt.show()
```
